# Manipulacin de data frames

```python
df = pd.DataFrame({
        'key': ['A', 'B', 'C', 'A', 'B', 'C'],
        'value1': range(6),
        'value2': np.random.randint(0, 100, 6)
    
})
```

```python
df
```

# Aggregate

```python
df.groupby('key').aggregate(['min', np.median, np.mean, max ])
```

```python
df.groupby('key').aggregate({
    'value1': 'min',
    'value2': 'max'
})
```

# Filter

```python
df.groupby('key').std()
```

```python
def filter_func(x): 
    return x['value2'].std()>5
```

```python
display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func)
```

# Transform

```python
df.groupby('key').transform(lambda x: -x.mean())
```

# Apply

```python
def norm_by_col2(x):
    x['value1']/= x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2) # Value 1 / Sum(Value2)
```

# Grficos desde data frames y exportacin

```python
years = [year for year in range(1900, 2020)]
deads = [(y +np.random.uniform(0,100)-1850) for y in years]
```

```python
df = pd.DataFrame({
    'year': years, 
    'deads': deads
})
```

```python
df
```

```python
df.plot(kind = 'scatter', x= "year", y= "deads", figsize=(10, 8))
```

```python
my_plot = df.plot(kind = 'scatter', x= "year", y= "deads", figsize=(10, 8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("../jupyter-python/data_frame_plot.png")
```
