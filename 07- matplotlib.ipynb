# Matplotlib

```python
%matplotlib inline 
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
```

```python
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
```

```python
plt.plot(x,y,"ro")
plt.axis([0,6,9,20]) #[x min, xmax, ymin, y max]
plt.show()
```

```python
import numpy as np
```

```python
data = np.arange(0.0, 10.0, 0.2 )
data
```

```python
plt.plot(data,data, "r--", data, data**2, "bs", data**3, "g^")
plt.show()
```

```python
plt.plot(x,y, linewidth=3.0) #linewidth : grosor de la linea
```

```python
line, = plt.plot(x,y, '-') #guarda el gráfico en una variable line
line.set_antialiased(False) #la imagen queda difuminada.
```

```python
lines = plt.plot(data, data, data,data**2)
plt.setp(lines, color= "r", linewidth = 2.0)
```

```python
lines = plt.plot(data, data, data,data**2) # dos gráficas .plot(x,y, (x, y^2))
plt.setp(lines, "color", "r", "linewidth", 2.0)
```
