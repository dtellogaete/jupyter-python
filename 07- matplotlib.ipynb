## Textos en los gráficos

### Histograma de frecuencias

```python
mu= 100
sigma = 20
x = mu +sigma*np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed = 1, facecolor = "g", alpha  =0.6)
plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$")
plt.text(40, 0.015, r'$\mu= 100, \ \sigma= 20$')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize =(10,6), dpi = 90 )  #Define la altura del gráfico, dpi = pixeles x pulgada
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0)

plt.annotate('Máximo Local', xy = (4*np.pi, 1), xytext = (15,1.5), 
             arrowprops = dict(facecolor = "black", shrink= 0.05))
plt.ylim(-2,2)
plt.show()
```

Pintar diferentes gráficos


```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0)&(y<1)]
y.sort()
x = np.arange(len(y))
```

```python
from matplotlib.ticker import NullFormatter
```

```python
plt.figure(figsize = (10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2) # Escala logaritmica
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(2,2,3) # Logaritmico simetrico
plt.plot(x,y -y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala Log Simétrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) # Anula formatos
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom =0.08, left = 0.1, right = 0.95, hspace = 0.25, wspace= 0.35)


plt.show()
```

### Cambios en los ejes de dibujos


```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True) # incluye ambos extremos
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure(figsize = (10,8)) 
plt.plot(x,S, color = "blue", linewidth =1.2, linestyle = "-", label = "seno")
plt.plot(x,C, color = "green", linewidth =1.2, linestyle = "-", label = "coseno")
plt.xlim(-4,4)
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
                       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # agrega separadores espaciando de -4 a 4
plt.yticks(np.linspace(-1, 1, 3, endpoint = True))

ax = plt.gca() # Modifica los ejes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.legend(loc = "upper left")

x0 = 2*np.pi/3
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 2.5, linestyle= "--") # Linea azul ---
plt.scatter([x0,], [np.sin(x0),], 50, color = "blue")
plt.annotate(r'$\sin(\frac{2\pi}{3})= \frac{\sqrt{3}}{2}$',
            xy= (x0, np.sin(x0)), xycoords = "data", 
            xytext = (+10,+30), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

for label in ax.get_xticklabels() +ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor = "white", edgecolor= "None", alpha = 0.6))

plt.show()
```
