## Textos en los grficos

### Histograma de frecuencias

```python
mu= 100
sigma = 20
x = mu +sigma*np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed = 1, facecolor = "g", alpha  =0.6)
plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$")
plt.text(40, 0.015, r'$\mu= 100, \ \sigma= 20$')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize =(10,6), dpi = 90 )  #Define la altura del grfico, dpi = pixeles x pulgada
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0)

plt.annotate('Mximo Local', xy = (4*np.pi, 1), xytext = (15,1.5), 
             arrowprops = dict(facecolor = "black", shrink= 0.05))
plt.ylim(-2,2)
plt.show()
```

Pintar diferentes grficos


```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0)&(y<1)]
y.sort()
x = np.arange(len(y))
```

```python
from matplotlib.ticker import NullFormatter
```

```python
plt.figure(figsize = (10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2) # Escala logaritmica
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(2,2,3) # Logaritmico simetrico
plt.plot(x,y -y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala Log Simtrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logstica")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) # Anula formatos
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom =0.08, left = 0.1, right = 0.95, hspace = 0.25, wspace= 0.35)


plt.show()
```
