```python
plt.figure(figsize =(15,8))
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('Total de nacimientos por a√±o')
plt.show()
```

```python

quartiles = np.percentile(births['births'],[25,50,75])
quartiles
```

```python
med = quartiles[1]
IQR = quartiles[2]-quartiles[0]
sig = 0.75*IQR
```

```python
births = births.query("(births > @med-5*@sig) & (births < @med +5*@sig)") #Los valores @ est√°n definidos anteriormente
```

```python
births.shape
```

```python
births['day'] = births['day'].astype(int) #Transformarlo en dato de tipo entero
```

```python
births.head()
```

```python
births.index = pd.to_datetime(10000*births.year+100*births.month+births.day, format = "%Y%m%d")
```

```python
births.head()
```

```python
births['dayofweek'] = births.index.dayofweek
```

```python
births.head()
```

```python
births.head()
```

```python
births.pivot_table('births', index= 'dayofweek', columns = 'decade', aggfunc= "mean")
```

```python
births.pivot_table('births', index= 'dayofweek', columns = 'decade', aggfunc= "mean").plot()
plt.gca().set_xticklabels(['Lun', 'Mar', 'Mier', 'Jue', 'Vie', 'Sab', 'Dom'])
plt.ylabel('Media de nacimientos por d√≠a')
plt.show()
```

```python
births_by_date = births.pivot_table('births', index=[births.index.month, births.index.day], 
                                   aggfunc='mean')
```

```python
births_by_date.head()
```

```python
births_by_date.head()
```

```python
births_by_date.index = [pd.datetime(2020, month, day) for (month, day) in births_by_date.index]
```

```python
births_by_date.head()
```

```python
plt.figure(figsize = (12,4))
births_by_date.plot()
plt.show()
```
